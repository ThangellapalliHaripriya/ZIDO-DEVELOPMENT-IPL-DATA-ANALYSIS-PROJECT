```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
matches = pd.read_csv("/content/matches.csv")
```

```python
deliveries = pd.read_csv("/content/deliveries.csv" ,on_bad_lines='skip')
```

```python
matches.head(20)
```

```python
deliveries.head(10)
```

```python
 matches.isnull().sum()
```

```python
matches['city'] = matches['city'].fillna(matches['city'].mode()[0])
matches['player_of_match'] = matches['player_of_match'].fillna('Unknown')
matches['winner'] = matches['winner'].fillna('No Result')
matches['result_margin'] = matches['result_margin'].fillna(0)
matches['target_runs'] = matches['target_runs'].fillna(0)
matches['target_overs'] = matches['target_overs'].fillna(0)
matches['method'] = matches['method'].fillna('Normal')
```

```python
deliveries.isnull().sum()
```

```python
deliveries['total_runs'] = deliveries['total_runs'].fillna(0)
```

```python
deliveries['extras_type'] = deliveries['extras_type'].fillna('None')
```

```python
deliveries['is_wicket'] = deliveries['is_wicket'].fillna(0)
```

```python
deliveries['player_dismissed'] = deliveries['player_dismissed'].fillna('None')
deliveries['dismissal_kind'] = deliveries['dismissal_kind'].fillna('None')
```

```python
deliveries['fielder'] = deliveries['fielder'].fillna('None')
```

```python
matches .isnull().sum()
matches.head(20)
```

```python
deliveries.isnull().sum()
deliveries.head(20)
```

Total Matches per Season

```python
matches_per_season = matches['season'].value_counts().sort_index()
sns.barplot(x=matches_per_season.index, y=matches_per_season.values,facecolor = 'g')
plt.title('Matches per Season')
plt.xlabel('Season')
plt.ylabel('Number of Matches')
plt.xticks(rotation=45)
plt.show()
```

Top Winning Teams

```python
top_teams = matches['winner'].value_counts().head(10)
sns.barplot(x=top_teams.values, y=top_teams.index)
plt.title('Top Winning Teams')
plt.xlabel('Wins')
plt.ylabel('Teams')
plt.show()
```

 Top Run Scorers


```python
top_batters = deliveries.groupby('batter')['batsman_runs'].sum().sort_values(ascending=False).head(10)
top_batters.plot(kind='barh', color='skyblue')
plt.title('Top Run Scorers')
plt.xlabel('Runs')
plt.ylabel('Players')
plt.show()
```



Top Wicket Takers

```python
wickets = deliveries[deliveries['is_wicket'] == 1]
top_bowlers = wickets['bowler'].value_counts().head(10)
top_bowlers.plot(kind='barh', color='salmon')
plt.title('Top Wicket Takers')
plt.xlabel('Wickets')
plt.ylabel('Bowlers')
plt.show()
```

Toss Decision Impact


```python
sns.countplot(data=matches, x='toss_decision', hue=matches['toss_winner'] == matches['winner'])
plt.title('Toss Decision vs Match Win')
plt.xlabel('Toss Decision')
plt.ylabel('Count')
plt.legend(['Lost', 'Won'])
plt.show()
```

Matches by Venue

```python
venue_counts = matches['venue'].value_counts().head(10)
sns.barplot(y=venue_counts.index, x=venue_counts.values, palette='viridis')
plt.title('Top Venues with Most Matches')
plt.xlabel('Number of Matches')
plt.ylabel('Venue')
plt.show()
```

Head-to-Head Team Wins

```python
# Grouping and counting wins
head_to_head = matches.groupby(['team1', 'team2'])['winner'].count().reset_index()

# Sorting top 10 matchups by number of wins
top_matches = head_to_head.sort_values('winner', ascending=False).head(10)

# Combine team1 and team2 names into one label
top_matches['matchup'] = top_matches['team1'] + " vs " + top_matches['team2']

# Plotting
plt.figure(figsize=(10, 6))
plt.barh(top_matches['matchup'], top_matches['winner'], color='skyblue')
plt.xlabel('Number of Wins')
plt.title('Top 10 Head-to-Head Team Matchups')
plt.gca().invert_yaxis()  # Highest bar on top
plt.tight_layout()
plt.show()
```
